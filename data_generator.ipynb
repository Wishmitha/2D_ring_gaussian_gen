```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean = [0, 0]
cov = [[1, 0], [0, 1]]
```

```python
x, y = np.random.multivariate_normal(mean, cov, 1000).T
plt.plot(x, y, '.')
plt.axis('equal')
plt.show()
```

Above is how we generate 2D symmetric Gaussian with mean [0,0]

In a ring like distribution the mean is not static as above. The mean of the distribution lies in a circle. We can use parametric form of a circle,

f(t) = (r.sin(t), r.cos(t))

to represent this mean.

```python
cov = [[1, 0], [0, 1]] #change the variances
N= 1000 # number of points to be generated
r = 50 # radius of the ring/circle
a,b = 0,0 # center of the ring/circle
```

```python
x = []
y = []
for i in range(N):
    t = np.random.uniform(0, 2*np.pi)
    mean = [a + r*np.sin(t), b + r*np.cos(t)]
    x.append(np.random.multivariate_normal(mean, cov).T[0])
    y.append(np.random.multivariate_normal(mean, cov).T[1])
plt.plot(x, y, '.')
plt.axis('equal')
plt.show()
```

```python
def gen_2D_ring_gaussian(N=1000, r1=50, r2=50, c1=0, c2=0, cov_mat=None):

    if cov_mat is not None:
        cov = cov_mat
    else:
        cov = [[1, 0], [0, 1]]
        
    x = []
    y = []
    for i in range(N):
        t = np.random.uniform(0, 2*np.pi)
        mean = [c1 + r1*np.sin(t), c2 + r2*np.cos(t)]
        x.append(np.random.multivariate_normal(mean, cov).T[0])
        y.append(np.random.multivariate_normal(mean, cov).T[1])

    return x,y
```

```python
x, y = gen_2D_ring_gaussian() #with default params
plt.plot(x, y, '.')
plt.axis('equal')
plt.show()
```

```python
x, y = gen_2D_ring_gaussian(N=100) # N=100 and other default params
plt.plot(x, y, '.')
plt.axis('equal')
plt.show()
```

```python
x, y = gen_2D_ring_gaussian(r1=50, r2=25) # r1=50 and r2=25 and other default params for a elliptical ring
plt.plot(x, y, '.')
plt.axis('equal')
plt.show()
```

```python
x, y = gen_2D_ring_gaussian(c1=25, c2=50) # making c1=25 and c2=50 and other default params
plt.plot(x, y, '.')
plt.axis('equal')
plt.show()
```

```python
x, y = gen_2D_ring_gaussian(cov_mat=[[50, 0], [0, 50]]) # Increasing the ring thickness by increasing the variance of covariance matrix -> variance is denoted diagonally in the covariance matrix
plt.plot(x, y, '.')
plt.axis('equal')
plt.show()
```

In below examples, we skew the default distribution by changing the covariance matrix

```python
x, y = gen_2D_ring_gaussian(cov_mat=[[1, 100], [1, 1]]) # skewing default distribution across vertical axis
plt.plot(x, y, '.')
plt.axis('equal')
plt.show()
```

```python
x, y = gen_2D_ring_gaussian(cov_mat=[[1, 1], [100, 1]]) # skewing default distribution across horizontal axis
plt.plot(x, y, '.')
plt.axis('equal')
plt.show()
```
