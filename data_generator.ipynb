```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mean = [0, 0]
cov = [[1, 0], [0, 1]]
```

```python
x, y = np.random.multivariate_normal(mean, cov, 1000).T
plt.plot(x, y, '.')
plt.axis('equal')
plt.show()
```

Above is how we generate 2D symmetric Gaussian with mean [0,0]

In a ring like distribution the mean is not static as above. The mean of the distribution lies in a circle. We can use parametric form of a circle,

f(t) = (r.sin(t), r.cos(t))

to represent this mean.

```python
cov = [[1, 0], [0, 1]] #change the variances
N= 1000 # number of points to be generated
r = 50 # radius of the ring/circle
a,b = 0,0 # center of the ring/circle
```

```python
x = []
y = []
for i in range(N):
    t = np.random.uniform(0, 2*np.pi)
    mean = [a + r*np.sin(t), b + r*np.cos(t)]
    x.append(np.random.multivariate_normal(mean, cov).T[0])
    y.append(np.random.multivariate_normal(mean, cov).T[1])
plt.plot(x, y, '.')
plt.axis('equal')
plt.show()
```
